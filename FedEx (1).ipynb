## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Copy the dataset for backup
FedEx_raw = FedEx.copy()
```

```python
# Clean column names: remove special characters and convert to snake_case
FedEx.columns = (
    FedEx.columns
    .str.strip()
    .str.lower()
    .str.replace(r'[^a-zA-Z0-9]+', '_', regex=True)  # This is key
    .str.strip('_')  # Remove leading/trailing underscores
)
```

```python
#View Updataed columns
print("Updated Columns Names:= ",FedEx.columns)
```

```python
#Convert relevent columns to datetime
data_cols=[
    "pq_first_sent_to_client_date",
    "po_sent_to_vendor_date",
    "scheduled_delivery_date",
    "delivered_to_client_date",
    "delivery_recorded_date"
]
for col in data_cols:
  FedEx[col]=pd.to_datetime(FedEx[col],errors='coerce')
```

```python
#converting into numeric columns
numeric_cols=[
    'line_item_quantity','line_item_quantity','pack_price','unit_price',
    'weight_kilograms','freight_cost_usd','line_item_insurance_usd'
]
for col in numeric_cols:
  FedEx[col]=pd.to_numeric(FedEx[col],errors='coerce')
```

```python
#check and drop duplicate value
print("Duplicate Rows: ",FedEx.duplicated().sum())
FedEx.drop_duplicates(inplace=True)
```

```python
#Check missing values
missing_value=FedEx.isnull().sum()
print("missing_values: \n",missing_value[missing_value>0])
```

```python
# Fill missing values for each column
FedEx['shipment_mode'].fillna(FedEx['shipment_mode'].mode()[0], inplace=True)
FedEx['dosage'].fillna('Unknown', inplace=True)
FedEx['weight_kilograms'].fillna(FedEx['weight_kilograms'].median(), inplace=True)
FedEx['freight_cost_usd'].fillna(0, inplace=True)
FedEx['line_item_insurance_usd'].fillna(0, inplace=True)
FedEx['pq_first_sent_to_client_date'].fillna(FedEx['pq_first_sent_to_client_date'].median(), inplace=True)
FedEx['po_sent_to_vendor_date'].fillna(FedEx['po_sent_to_vendor_date'].median(), inplace=True)
```

```python
# For categorical variables, fill with mode
cat_cols = FedEx.select_dtypes(include='object').columns
for col in cat_cols:
    FedEx[col] = FedEx[col].fillna(FedEx[col].mode()[0])
```

```python
# Final check
print("Data types after wrangling:\n", FedEx.dtypes)
```

#### Chart - 1

```python
# Filter top 10 countries based on frequency
top_countries = FedEx['country'].value_counts().nlargest(10).index
filtered_data = FedEx[FedEx['country'].isin(top_countries)]

# Create count plot
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_data, x='shipment_mode', hue='country')
plt.title('Shipment Mode Distribution for Top 10 Countries')
plt.xlabel('Shipment Mode')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

The grouped bar chart was selected because it is ideal for comparing categorical data across multiple groups. In this case, it allows for a clear comparison of different shipment modes (Air, Truck, Air Charter, and Ocean) across the top 10 countries with the highest number of shipments.

This chart makes it easy to observe variations in shipment preferences for each country side by side, helping to uncover patterns and trends that may not be as visible in other chart types like pie charts or single bar plots. Its visual clarity and ability to display multiple dimensions at once make it the best fit for this analysis.



##### 2. What is/are the insight(s) found from the chart?

The chart reveals that Air is the dominant shipment mode for most of the top 10 countries, especially in Côte d'Ivoire, Vietnam, Nigeria, and Haiti, indicating a preference for faster delivery services. South Africa, however, shows a significantly high reliance on Truck shipments, suggesting well-developed road logistics. Nigeria also stands out with the highest usage of Air Charter, which may imply urgent or high-priority shipments. Ocean shipping is used in moderate quantities by South Africa and Haiti, possibly due to their access to seaports. Overall, the shipment mode distribution varies across countries, likely influenced by geographical, infrastructural, and economic factors.

#### Chart - 3

```python
# Calculate top 10 countries with highest average delivery delay
avg_delay = FedEx.groupby('country')['delivery_delay'].mean().sort_values(ascending=False).head(10).index

# Filter dataset for these top 10
top_delay_data = FedEx[FedEx['country'].isin(avg_delay)]

# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=top_delay_data, x='country', y='delivery_delay', palette='coolwarm')
plt.title('Top 10 Countries by Delivery Delay')
plt.xlabel('Country')
plt.ylabel('Delivery Delay (Days)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

The boxplot was specifically chosen for this visualization because it is ideal for showing the distribution of a continuous variable—such as delivery delay—across different categorical groups, in this case, the top 10 countries. It provides a concise yet comprehensive summary of each group’s data, including the median, interquartile range (IQR), and outliers. This makes it easy to compare the consistency and central tendency of delivery times across countries. Additionally, since delivery delay data may include extreme values or skewed distributions, a boxplot is particularly useful as it does not assume any specific distribution and can highlight irregularities effectively.

##### 2. What is/are the insight(s) found from the chart?

From the chart, we gain several insights. Most notably, Congo, DRC has the highest median delivery delay, a large IQR, and a significant number of outliers, indicating frequent and unpredictable delays. This suggests serious inefficiencies in the delivery process within that country. On the other hand, countries like Senegal, Lesotho, Angola, Liberia, and Guinea show narrow IQRs and medians close to zero, reflecting stable and timely deliveries. Kenya and Benin demonstrate moderate delays with a few high outliers, indicating occasional but impactful issues. Interestingly, some countries like Kenya and Togo also show negative delivery delays, which may suggest deliveries being made ahead of schedule—possibly pointing to inaccurate delivery time estimates.

The insights from this chart can indeed lead to positive business impact. Identifying problematic regions like Congo, DRC allows the business to take targeted actions to investigate and resolve root causes such as poor infrastructure, customs delays, or inefficient logistics operations. ##### 1. Why did you pick the specific chart?

 Scatter plot was chosen to explore the relationship between two numerical variables, delivery delay and freight cost, while also incorporating a categorical variable, shipment mode, through color coding. This allows us to see if there's any correlation between how much is spent on freight and the resulting delivery time, and whether this relationship differs based on the shipment method used.

##### 2. What is/are the insight(s) found from the chart?

The scatter plot analysis of freight cost versus delivery delay, segmented by shipment mode, reveals the following:

Limited Cost-Time Correlation: A weak linear relationship exists between freight expenditure and delivery timeliness across all shipment methods. Higher freight costs do not consistently correlate with reduced delivery delays.
Modal Performance Clusters: Distinct clusters emerge based on shipment mode:
Truck: Characterized by lower freight costs and a broad distribution of delivery delays, encompassing both significant early arrivals and moderate lateness.
Air: Exhibits a wider range of freight costs, with delivery delays predominantly centered around the scheduled time, though instances of both early and late deliveries are present.
Air Charter: Associated with higher freight costs and a more dispersed range of delivery delays, including the most extreme early and late occurrences.
Ocean: Typically involves lower to mid-range freight costs and a wide spectrum of delivery delays, with notable instances of both early and late arrivals.
Inefficient High-Cost Shipments: Several outliers indicate shipments incurring substantial freight costs yet experiencing considerable positive or negative delivery delays, suggesting potential inefficiencies in premium shipping services.
